```python
# クイックソート
# 値を格納したリスト（numbers）
numbers = [6, 19, 30, 22, 19, 1, 14]

if numbers[0] != max(numbers):
    # 左端の値を基準値とする
    i = numbers[0]
    # 左端の値がリスト内で最大値を持つ場合は、要素1の値を基準値とする
else:
    i = numbers[1]

for i in range(0,len(numbers)):
    for j in range(0, len(numbers)):
        # 同じインデックス番号のものでは比較しない
        if not numbers[i] == numbers[j]:
            # 基準値の値が右隣の値よりも大きい
            if numbers[i] > numbers[j]:
                # 基準値の左側に、右隣の要素を移動させる
                numbers[j], numbers[i] = numbers[i], numbers[j]
            # 基準値の値が右隣の値よりも小さい
            elif numbers[i] < numbers[j]:
                # 右隣の要素の移動はなし
                numbers[i], numbers[j] = numbers[i], numbers[j]
```
