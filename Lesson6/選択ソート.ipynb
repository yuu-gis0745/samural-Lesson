```python
# 選択ソート
numbers = [6, 19, 30, 22, 15, 1]
# numbersのリストの長さを取得
l = len(numbers)

# 0～リストの長さ-1の要素をループ変数jに代入（リストの長さは5であるため、0～4までの要素を代入）
for j in range(0, l - 1):
    # j～lまでの範囲で最小値のインデックス番号を取得し、iに代入する
    # ループするごとに、最小値のインデックスを調べる範囲をjとlで調整し、左端の値を「最小値」と判断しないようにする
    i = numbers.index(min(numbers[j: l]))
    # 最小値とリストの左側の値を比較する
    if numbers[i] < numbers[j]:
        # 最小値とリストの左側の値を入れ替える
        numbers[j], numbers[i] = numbers[i], numbers[j]

print(numbers)
```
