```python
# バブルソート
numbers = [6, 19, 30, 22, 15, 1]
# numbersのリストの長さを取得
l = len(numbers)

# 外側のループでは、numbers[0]～numbers[l-1](インデックス5まで)の処理を繰り返す
for i in range(l - 1):
        # 内側のループでは、numbers[l-1]インデックス5(リストの最後)～numbers[0](リストの最初)の間で処理を繰り返す
        for j in range(l -1 , 0, -1):
        # リストの隣り合う前後の要素を比較する
            if numbers[j] < numbers[j - 1]:
                numbers[j - 1], numbers[j] = numbers[j], numbers[j - 1]

print(numbers)
```
