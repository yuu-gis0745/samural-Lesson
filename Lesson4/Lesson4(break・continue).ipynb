```python
#ランダムな整数を利用するため、記述が必要です。
import random

for i in range(1, 11):
    #変数numに1～30までのランダムな整数を代入する
    num = random.randint(1, 30)

    print(f"{i}回目の結果は{num}です。")

    #変数numの値が20であれば、break分で繰り返し処理を強制終了する
    if num == 20:
        print("20が出たので繰り返し処理を強制終了します。")
        break
    print("終了しました")
```

```python
#ランダムな整数を利用するため、記述が必要です。
import random

for i in range(1, 11):
    #変数numに1～30までのランダムな整数を代入する
    num = random.randint(1, 30)

    print(f"{i}回目の結果は{num}です。")

    #変数numの値が20であれば、continue分で処理から抜け出す
    if num == 20:
        print("20が出たので処理から抜け出します。")
        continue
    print("終了しました")
```
